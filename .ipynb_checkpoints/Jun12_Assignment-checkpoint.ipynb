```python
import pandas as pd
from pandas import Series, DataFrame
import numpy as np
```

```python
df=pd.read_csv('titanic.csv')
df.head()
```

```python
age_mean=np.round(df.Age.mean())
age_mean
```

```python
df.Age.fillna(age_mean,inplace=True)
df.head()
```

```python
bins=[0,15,60,120]
```

```python
df['Agerange']=pd.cut(df.Age,bins)
df.head()
```

```python
grouped=df.groupby(['Pclass','Agerange'],)
```

```python
functions = [('합계','count'),('생존자','sum'),('생존율','mean')]
result = grouped['Survived'].agg(functions)
result
```
