```python
from pandas import Series, DataFrame

import pandas as pd
import numpy as np
from datetime import datetime
```

```python
price=pd.read_pickle('yahoo_price.pkl')
volume=pd.read_pickle('yahoo_volume.pkl')
```

```python
display(price.head())
display(volume.head()) 
```

```python
m_price=price.set_index([price.index.strftime('%Y'), price.index.strftime('%m-%d')])
m_volume=volume.set_index([price.index.strftime('%Y'), price.index.strftime('%m-%d')])
```

```python
m_price.index.names = ['year', 'month-day']
m_volume.index.names = ['year', 'month-day']
```

```python
display(m_price.head(10))
display(m_volume.tail(10))
```

```python
display(m_price.pct_change().head(10))
display(m_volume.pct_change().tail(10))
```

```python
m_price=pd.concat([m_price,m_price.pct_change()],axis=1)
m_price.head()
```
